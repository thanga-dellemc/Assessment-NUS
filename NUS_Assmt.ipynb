```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/thanga-dellemc/Assessment-NUS/main/Sheet1.csv')
df.head()
data = df['Scam Call']

ydf = pd.DataFrame(data)

# Create a dictionary for mapping values
mapping = {'Not Scam': 0, 'Scam': 1, 'Unlikely': 2}

# Replace values in the column using the mapping dictionary
ydf['Scam Call'] = ydf['Scam Call'].replace(mapping)

# Print the DataFrame
#

# data separation of y and x axis
y = ydf['Scam Call']
x = df[['Call Frequency', 'Previous Contact Count']]

ydf['Scam Call'] = ydf['Scam Call'].astype(int)
y = ydf['Scam Call']


# split data for prediction and testing the model
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
x_train
x_test

print(x_train,y_train)

df_merged = pd.concat([x_train, y_train], axis=1)

print(df_merged.corr()["Scam Call"])
import seaborn as sns
sns.lmplot(x="Previous Contact Count", y="Call Frequency", data=df_merged, fit_reg=True, ci=None)


# model building linear regression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
predictors = ["Call Frequency", "Previous Contact Count"]
target = "Scam Call"
lr.fit(df_merged[predictors], df_merged[target])
LinearRegression()


```
